```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#read in the data

df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df['rating'].fillna('Not-rated', inplace =True)
```

```python
budget_medians = df.groupby(['genre', 'year'])['budget'].transform('median')
budget_medians
df['budget'].fillna(budget_medians, inplace = True)
df['budget'].isnull().sum()
df['budget'].fillna(df['budget'].median(), inplace = True)
df['budget'].isnull().sum()

gross_medians = df.groupby(['genre', 'year'])['gross'].transform('median')
df['gross'].fillna(gross_medians, inplace = True)
df['gross'].isnull().sum()
df['gross'].fillna(df['gross'].median(), inplace = True)
df['gross'].isnull().sum()

df['company'].fillna('Unknown', inplace = True)

```

```python
df.isna().sum()
df.dropna
```

```python
# prompt: drop all missing values

df.dropna(inplace=True)
```

```python
# prompt: findout duplicate rows

duplicate_rows = df[df.duplicated()]
print("Duplicate Rows:")
duplicate_rows
```

```python
df['budget'] = df['budget'].astype(int)
df['gross'] = df['gross'].astype(int)
df.dtypes
```

```python
# prompt: split the  released column

# Assuming 'released' column contains date strings in a format like 'YYYY-MM-DD'
df[['year_released', 'month_released', 'day_released']] = df['released'].str.split('-', expand=True)

# If you want to handle different date formats, you might need more sophisticated parsing using libraries like dateutil.parser.
# Example using dateutil.parser:
# from dateutil import parser
# df['parsed_date'] = df['released'].apply(lambda x: parser.parse(x))
# df['year_released'] = df['parsed_date'].dt.year
# df['month_released'] = df['parsed_date'].dt.month
# df['day_released'] = df['parsed_date'].dt.day
```

```python
df[['released_date', 'released_country']] = df['released'].str.split('(', expand=True)
df.head()
```

```python
df['released_date'] = pd.to_datetime(df['released_date'], infer_datetime_format= True, errors= 'coerce')
df.head()

df['released_year'] = df['released_date'].dt.year
df.head()
df['released_year'].fillna(df['year'], inplace = True)
df.head()
```

```python
df['released_year'] = df['released_year'].astype(int)
df.head()
```

```python
sns.scatterplot(data = df, x = 'budget', y = 'gross')
plt.show()
```

```python
sns.regplot(data=df, x='budget', y='gross', line_kws={'color': 'Red'})
plt.show()
```

```python
# Select only numerical features for correlation calculation
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
